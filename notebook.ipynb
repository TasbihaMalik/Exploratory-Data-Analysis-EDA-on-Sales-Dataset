# 📚 Step 1: Import Required Libraries

This section imports all Python libraries required for data analysis and visualization.

These libraries will be used for:

- Reading Excel files
- Data manipulation
- Statistical analysis
- Data visualization

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

# 📂 Step 2: Load the Dataset

The cleaned sales dataset is loaded from the **Data** folder using Pandas.

Once loaded, the first few rows of the dataset will be displayed to verify that the data has been imported successfully.

```python
df = pd.read_excel("Data/Dataset for Data Analytics.xlsx")

df.head()
```

# 🔍 Step 3: Dataset Overview

Before performing any analysis, it is important to understand the dataset.

This section examines:

- Number of rows and columns
- Column names
- Data types
- Missing values

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

### Observation

The dataset has been loaded successfully.

The output shows the total number of rows, columns, data types, and confirms whether the dataset contains any missing values.

# 📈 Step 4: Descriptive Statistics

Descriptive statistics provide a summary of the numerical columns in the dataset.

The following statistics will be calculated:

- Mean
- Median
- Minimum
- Maximum
- Standard Deviation
- Quartiles

```python
df.describe()
```

```python
summary = df.describe()

summary.to_csv("reports/summary_statistics.csv")
```

### Observation

The summary statistics provide a quick overview of the dataset.

These values help understand the average, spread, and distribution of numerical variables.

# 🧹 Step 5: Missing Value Analysis

Missing values can negatively affect analysis.

This section checks whether the dataset contains any missing values.

```python
df.isnull().sum()
```

### Observation

The dataset contains very few or no missing values, indicating that it is suitable for further analysis.

# 🔗 Step 6: Correlation Analysis

Correlation measures the strength of the relationship between numerical variables.

The correlation matrix will be visualized using a heatmap.

```python
correlation = df.corr(numeric_only=True)

correlation
```

```python
correlation.to_csv("reports/correlation_matrix.csv")
```

# 🌡 Correlation Heatmap

The heatmap visually represents relationships between numerical variables.

Higher values indicate stronger positive relationships.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.savefig("visualizations/heatmap.png")

plt.show()
```

### Observation

The heatmap shows how strongly different numerical variables are related.

Strong positive or negative correlations can provide useful business insights.

# 📊 Step 7: Distribution Analysis

Histograms help understand how numerical values are distributed.

This analysis identifies:

- Skewness
- Peaks
- Spread

```python
plt.figure(figsize=(8,5))

sns.histplot(df["TotalPrice"], bins=30)

plt.title("Distribution of Total Price")

plt.savefig("visualizations/totalprice_distribution.png")

plt.show()
```

### Observation

The histogram illustrates the frequency distribution of Total Price.

Most sales are concentrated within a specific price range, while a smaller number of high-value transactions create a right-skewed distribution.

# 📦 Step 8: Outlier Detection

A boxplot is used to identify extreme values within the Unit Price column.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["UnitPrice"])

plt.title("Unit Price Outliers")

plt.savefig("visualizations/unitprice_boxplot.png")

plt.show()
```

### Observation

The boxplot indicates the presence of outliers.

These observations may represent premium products or unusual transactions.

# 💳 Step 9: Payment Method Analysis

This section identifies the most frequently used payment methods.

```python
df["PaymentMethod"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Payment Methods")

plt.savefig("visualizations/payment_method.png")

plt.show()
```

# 📦 Step 10: Order Status Analysis

This section analyzes customer order statuses.

```python
df["OrderStatus"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Order Status")

plt.savefig("visualizations/order_status.png")

plt.show()
```

# 🛒 Step 11: Product Performance Analysis

The objective is to identify products generating the highest revenue.

```python
product_sales = df.groupby("Product")["TotalPrice"].sum()

product_sales.sort_values(ascending=False)
```

```python
product_sales.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Revenue by Product")

plt.savefig("visualizations/product_sales.png")

plt.show()
```

# 📅 Step 12: Monthly Sales Trend

Monthly sales trends help identify seasonal business patterns.

```python
df["Month"] = df["Date"].dt.month_name()

monthly_sales = df.groupby("Month")["TotalPrice"].sum()

monthly_sales.plot(
    kind="line",
    marker="o",
    figsize=(10,5)
)

plt.title("Monthly Sales")

plt.savefig("visualizations/monthly_sales.png")

plt.show()
```

# 👥 Step 13: Top Customers

This section identifies the customers who generated the highest revenue.

```python
top_customers = (
    df.groupby("CustomerID")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers
```
